```python
from flwr_datasets import FederatedDataset
from flwr_datasets.partitioner import IidPartitioner

num_partitions = 5  # or however many clients you want
partitioner = IidPartitioner(num_partitions=num_partitions)
fds = FederatedDataset(
    dataset="khoaguin/pima-indians-diabetes-database",
    partitioners={"train": partitioner},
)
```

```python
# To load the partition for client 0:
partition_0 = fds.load_partition(0, "train").with_format("pandas")[:]

# For client 1:
partition_1 = fds.load_partition(1, "train").with_format("pandas")[:]
```
